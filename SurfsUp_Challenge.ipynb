```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# Import the sqlalchemy extract function.
from sqlalchemy import extract

# 1. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==6)
```

```python
# 2. Convert the June temperatures to a list.
June_results = session.query(Measurement.date, Measurement.tobs).filter(extract('month',Measurement.date)==6).all()
print(June_results)
```

```python
# 3. Create a DataFrame from the list of temperatures for the month of June. 
June_df = pd.DataFrame(June_results, columns=['date','June Temps'])
June_df.set_index(June_df['date'], inplace=True)

# convert df to string  and set index = False so that it prints the data with date and temperature cols.
print(June_df.to_string(index=False))

```

```python
# 4. Calculate and print out the summary statistics for the June temperature DataFrame.
June_df.describe()
```

### D3: Determine the Rainfall  Statistics for June and December


```python
June_rain_results = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
df_rain_june = pd.DataFrame(June_rain_results, columns=['date','June Rain'])
df_rain_june.describe()
```

```python
Dec_rain_results = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
df_rain_dec = pd.DataFrame(Dec_rain_results, columns=['date','Dec Rain'])
df_rain_dec.describe()
```

```python
#Drop the null values:
df_rain_june = df_rain_june.dropna()
df_rain_dec = df_rain_dec.dropna()

#Style
plt.style.use('seaborn-whitegrid')

#Plot the Box & Whiskers
x_labels = ["June Rain", "Dec Rain"]
temps = [df_rain_june["June Rain"], df_rain_dec["Dec Rain"]]
fig, ax = plt.subplots(figsize=(10, 7))
ax.boxplot(temps, labels=x_labels)


#Ticks & Labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)

ax.set_yticks(np.arange(0, 7, step=0.6))
ax.set_title('Rainfall Data',fontsize=18)
ax.set_ylabel('Rainfall (inches)',fontsize=14)
ax.set_xlabel("Month",fontsize=14)
```

## Average rainfall in Oahu


```python
#December Rain group by Year
#Set index to date
df_index_dec_rain = df_rain_dec.set_index(df_rain_dec['date'], inplace=False)

#Sort the dates
df_index_dec_rain = df_index_dec_rain.sort_index()

#Set the index date to the date datatype
df_index_dec_rain.index = pd.to_datetime(df_index_dec_rain.index)

#Group by year
df_rain_dec_index_year = df_index_dec_rain.resample("Y").mean()
df_rain_dec_index_year
```

```python
#June Rain by Year
df_index_jun_rain = df_rain_june.set_index(df_rain_june['date'], inplace=False)

#Sort the dates
df_index_jun_rain = df_index_jun_rain.sort_index()

#Set the index date to the date datatype
df_index_jun_rain.index = pd.to_datetime(df_index_jun_rain.index)

#Group by year
df_rain_jun_index_year = df_index_jun_rain.resample("Y").mean()
df_rain_jun_index_year
```

```python
#June and December - graph - by year
#Style
style.use('ggplot')

#Plot the graph
ax = df_rain_jun_index_year.plot(c='green', linewidth=1)
df_rain_dec_index_year.plot(ax=ax, c='red', figsize=(12, 5), linewidth=1)

#Labels
ax.set_xlabel('Year')
ax.set_ylabel("Average Rainfall in inches")
ax.set_title("Average Rainfall in Oahu for June and Decembr")

#Ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)

#Plot the legend
lgnd = plt.legend(fontsize="10", mode="Expanded",
         loc="best", title="Legend")
lgnd.get_title().set_fontsize(11)
plt.show()
```
